```python
df = pd.DataFrame(np.arange(12).reshape(3, 4),
                  columns=['A', 'B_X', 'B', 'C'])

print('original df')
print(df)

print('\nselect using a list')
print(df.select(['A','B_X']))

print('\nselect using a regex')
print(df.select('^B|C'))

print('\nselect using a tuple of list and  regex')
print(df.select(['A','B_X'],'^B|C'))
```
